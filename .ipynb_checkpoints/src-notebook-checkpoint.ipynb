```python
### Setting training data list ### Updated ###


training_data = []

for img_name in df.head(10)['Image Index']:
    object = bucket.Object(key)
    tmp = tempfile.NamedTemporaryFile()
    class_num = indexed_w_all.loc[img_name]['cardiomegaly']
    with open(tmp.name, 'wb') as f:
        object.download_fileobj(f)
        
        # plotting full image
        img_array=cv2.imread(tmp.name)
        
        #plotting scaled down image
        IMG_SIZE = 10
        
        new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
        avg_array = []
        for num in new_array:
            for nums in num:
                
                avg_array.append(np.mean(nums))
    
        training_data.append([img_name, avg_array, class_num])
            
```

```python
### Creating image processing dataframe ###

image_df = pd.DataFrame(training_data)
image_df = image_df.rename(columns={0: "img_name", 1: "avg_array", 2: "class_num"})
```

```python
### Creating dataframe with image names and categroization ###

smol_df = image_df.apply(pd.Series)
len(smol_df)
```

```python
### Creating dataframe with all image array data ###

matrix_df = image_df.avg_array.apply(pd.Series)
len(matrix_df)
```

```python
### Checking dataframe ###

matrix_df.head(2)
```

```python
### Concating the dataframes on index ###

df_merged = pd.concat([smol_df, matrix_df], axis=1, sort=False)
```

```python
### Dropping unnecessary column ###

df_merged = df_merged.drop('avg_array', axis = 1)
```

```python
### Checking concatenated dataframe ###

df_merged.head(2)
```

```python
# Creating X, y data ###

y = df_merged['class_num']
X = df_merged.drop(['img_name', 'class_num'], axis = 1)
```

```python
### Train test split the data ###

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=69)
    
```

```python
### Creating an SVM model ###

from sklearn import svm

clf = svm.SVC()
clf.fit(X_train, y_train)
svm.SVC()
```

```python
### Prediction & Accuracy ###

predicted = clf.predict(X_test)
print(accuracy_score(y_test, predicted))
```

```python
### Attempt at object oriented programming ###

class Rads_Reader:
    def __init__(self, bucket):
        self.bucket = bucket
    
    def Classification_df(self):
        ### AWS S3 Setup ###
        client = boto3.client('s3') #low-level functional API
        s3 = client
        resource = boto3.resource('s3') 


        ### Adding column to dataframe for cardiomegaly ###

        obj = client.get_object(Bucket = 'rads-reader', Key = 'data/categorical.csv')
        categorical_df = pd.read_csv(obj['Body'])
        categorical_df['cardiomegaly'] = [1 if 'Cardio' in x else 0 for x in categorical_df['Finding Labels']]

        ### Create condensed version of dataframe ###
        df = categorical_df[['Image Index','Finding Labels','cardiomegaly']]
        return df
        
    def Indexed_df(self):
        indexed_w_all = self.Classification_df.set_index('Image Index')
        return indexed_w_all
        
    def Training_data(self):
        s3 = boto3.resource('s3')
        bucket = s3.Bucket(bucket_name)
        response = s3.list_objects_v2(Bucket=bucket_name)
        training_data = []


        for obj in response['Contents'][3:]:

            key=obj['Key']
            img_name = key.lstrip('data/images/')


            object = bucket.Object(key)
            tmp = tempfile.NamedTemporaryFile()
            class_num = indexed_w_all.loc[img_name]['cardiomegaly']
            with open(tmp.name, 'wb') as f:
                object.download_fileobj(f)

                # plotting full image
                img_array=cv2.imread(tmp.name)

                #plotting scaled down image
                IMG_SIZE = 100

                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                avg_array = []
                for num in new_array:
                    for nums in num:

                        avg_array.append(np.mean(nums))

                training_data.append([img_name, avg_array, class_num])
        return training_data


```
